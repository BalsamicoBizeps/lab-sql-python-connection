```python
import pandas as pd
import numpy as np
import pymysql
from sqlalchemy import create_engine
import getpass  # To get the password without showing the input
password = getpass.getpass()
```

```python
sk = "sakila"
connection_string = 'mysql+pymysql://root:' + password + '@localhost/'+ sk
engine = create_engine(connection_string)
engine
```

```python
from sqlalchemy import text
```

```python
def rentals_month(engine, month, year):
    with engine.connect() as connection:
        txt = f'''SELECT rental_id, rental_date, inventory_id, customer_id, return_date, staff_id
        FROM rental
        WHERE rental_date LIKE "{year}-%{month}-%";
        '''
        query = text(txt)
        result = connection.execute(query)
        return pd.DataFrame(result.all())
```

```python
rentals_month(engine, 5, 2005)
```

```python
def rental_count_month(df, month, year):
    df = df.groupby("customer_id").agg({"rental_id":"count"}).reset_index()
    df.rename(columns={'rental_id': f"rentals_{month}_{year}"}, inplace=True) 
    return df
  
```

```python
rental_count_month(rentals_month(engine, 5, 2005), 5, 2005)
```

```python
def compare_rentals(df1, df2):
    df_merged = df1.merge(df2, how="inner", on="customer_id")
    df_merged.reset_index(drop=True, inplace=True)
    df_merged["difference"] = df_merged.iloc[:, 1] - df_merged.iloc[:, 2]
    return df_merged
```

```python
df1 = rental_count_month(rentals_month(engine, 5, 2005), 5, 2005)
df2 = rental_count_month(rentals_month(engine, 6, 2005), 6, 2005)

compare_rentals(df1, df2)
```
